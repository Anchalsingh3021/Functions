1. What is the difference between a function and a method in Python?
  - A function is a piece of code that you define using the def keyword, and it can be called directly by its name. It doesn't need to be tied to any object or class. For example, you can have a function that adds two numbers.

     On the other hand, a method is a function that is defined inside a class and is tied to an object (an instance of that class). Methods can operate on the data (attributes) that belongs to the object. When you call a method, you do so through an object or instance of the class, like object.method().


2.  Explain the concept of function arguments and parameters in Python.

  -  In Python, parameters and arguments are two key concepts used in functions. A parameter is a variable that is defined in the function's declaration. It acts as a placeholder that will receive the value when the function is called. For example, in the function def add(a, b):, a and b are parameters.

     Arguments are the actual values you pass to the function when calling it. For example, in add(2, 3), the numbers 2 and 3 are the arguments passed to the parameters a and b.

     In short, parameters are like the "boxes" inside the function that expect values, while arguments are the "things" you put in those boxes when you call the function. You can also use default parameters or keyword arguments to make functions more flexible.

5.  What are iterators in Python and how do they differ from iterables?

   - In Python, iterators and iterables are related concepts but serve different purposes.

     1. An iterable is any object that can be looped over, like a list, tuple, or string. It defines a __iter__() method, which returns an iterator object.

      2. An iterator is an object that represents a stream of data. It is what you get when you call iter() on an iterable. An iterator has a __next__() method, which returns the next item in the sequence each time it's called. When there are no more items to return, it raises a StopIteration exception.

6. Explain the concept of generators in Python and how they are defined.

   - Generators in Python are special types of iterators that allow you to iterate over a sequence of values one at a time, without storing the entire sequence in memory. This makes them very memory-efficient. Generators are defined using a function with the `yield` keyword instead of `return`, or through a generator expression, which looks like a list comprehension but uses parentheses.

     When a generator function is called, it doesn’t execute immediately; instead, it returns a generator object. Each time you request the next value (using `next()`), the function resumes from where it last left off, making it easy to handle large datasets or infinite sequences.

     Generators are ideal for situations where you only need to process one item at a time and want to save memory.


7. What are the advantages of using generators over regular functions?

   - Generators offer several advantages over regular functions, particularly when dealing with large data sets or complex iterations. Here are some key benefits:

  1. Memory Efficiency: Generators produce items one at a time, which means they don’t store the entire sequence in memory. This makes them much more memory-efficient compared to regular functions that return large lists or other data structures.

  2. Lazy Evaluation: Generators use lazy evaluation, meaning they generate values only when needed, rather than all at once. This can improve performance, especially when working with large datasets, as the program doesn't need to compute and store everything upfront.

  3. Better Performance: Since generators yield values on the fly, they can be faster in some cases compared to functions that generate and return all data at once. You can start processing values immediately without waiting for the entire collection to be built.

  4. Simpler Code: Using generators can make code more readable and concise. You can avoid the need for manual tracking of state and the use of complex loops or conditionals.

8.  What is a lambda function in Python and when is it typically used?

  - A lambda function in Python is a small, simple function that you can create on the fly, without needing to give it a name. It's defined using the lambda keyword, followed by the arguments it takes, a colon, and then an expression to evaluate. Unlike regular functions, which are defined with def, lambda functions are more compact and are usually written in a single line.

  we'll typically use a lambda function when we need a quick, short function for something simple, without the need to formally define it. For example, it’s often used in functions like map(), filter(), or sorted() when we want to perform a quick operation without writing a whole function.

  For example, if we wanted to add two numbers, you could use a lambda function like this:


    add = lambda x, y: x + y
    print(add(2, 3))  # Output: 5


9.  Explain the purpose and usage of the `map()` function in Python.

  - The map() function in Python allows you to apply a given function to every item in an iterable (such as a list, tuple, or string) without needing to write a loop. It simplifies your code and makes it more concise when you want to perform the same operation on all elements of an iterable.

   To use map(), we provide two arguments:

     1. A function that defines the operation to apply to each item.
     2. An iterable (like a list or tuple) to which the function will be applied.

   The result of map() is an iterator, which can be converted into a list or other collection to see the final output.

     map() is typically used when you need to perform the same transformation on each element in a collection, making your code more readable and efficient compared to using a loop.

10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

  - In Python, the `map()`, `reduce()`, and `filter()` functions are all used for processing iterables, but they serve different purposes and work in unique ways.

     1. **map()**
 The `map()` function applies a given function to each item in an iterable (like a list or tuple) and returns an iterator that yields the results.
 It’s used when you want to transform every element in an iterable using the same function.
 It returns an iterator containing the transformed values.

     2. **reduce()**
The `reduce()` function, found in the `functools` module, applies a given function cumulatively to the items in an iterable, from left to right, reducing the iterable to a single value.
 It's useful when you want to perform a cumulative operation, like summing or multiplying all elements in a list.
 It returns a single value, which is the result of applying the function to all items.

     3. **filter()**
 The `filter()` function applies a given function to each item in an iterable and filters out those that don't meet the condition defined by the function.
 It’s used when you want to filter elements based on some condition, such as keeping only even numbers or non-empty strings.
It returns an iterator containing only the items that satisfy the condition.

```python
# 1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example usage:
numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(numbers)
print(result)
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

# Example usage:
for number in range(1, 201):
    if is_prime(number):
        print(number, "is prime")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        fib_num = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return fib_num

# Example usage:
fib = FibonacciIterator(10)  # Generate first 10 Fibonacci numbers
for number in fib:
    print(number)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in powers_of_two(5):
    print(power)
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# List of tuples
tuples_list = [(1, 3), (4, 1), (2, 5), (3, 2)]

# Sort the list based on the second element of each tuple
sorted_list = sorted(tuples_list, key=lambda x: x[1])

# Output the sorted list
print(sorted_list)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.


celsius_temperatures = [0, 20, 30, 40, 100]

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Use map() to apply the conversion to each temperature in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print(fahrenheit_temperatures)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def is_not_vowel(char):
    return char.lower() not in 'aeiou'

input_string = "Hello, World!"

# Use filter() to remove vowels
result = ''.join(filter(is_not_vowel, input_string))

print(result)
```
